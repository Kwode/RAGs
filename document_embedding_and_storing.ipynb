```python
from langchain_openai import ChatOpenAI
from langchain_text_splitters import RecursiveCharacterTextSplitter, MarkdownHeaderTextSplitter
from langchain_classic.document_loaders import PyPDFLoader, Docx2txtLoader
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
from langchain_community.vectorstores import Chroma
from langchain_classic.docstore.document import Document
import copy
```

```python
load_dotenv()
```

```python
#load document
pdf_loader = Docx2txtLoader('Introduction_to_Data_and_Data_Science - Copy.docx')
docx = pdf_loader.load()

#copy pdf
docx_copy = copy.deepcopy(docx)

#remove new line characters
for i in docx_copy:
    i.page_content = ' '.join(i.page_content.split())
```

```python
#split document
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap =50, separators=['.'])

docx_split = text_splitter.split_documents(docx_copy)
docx_split[0]
```

## Text Embedding

```python
embeddings = OpenAIEmbeddings(model='text-embedding-3-small')
```

## Vector Storing

```python
vectorstore = Chroma()

vectorstore.from_documents(documents= docx_split, embedding= embeddings, persist_directory='./vector_db')
```

```python
vectorstore_from_dir = Chroma(persist_directory='./vector_db', embedding_function= embeddings)
```

```python
add_doc = Document(metadata={'source': 'Introduction_to_Data_and_Data_Science - Copy.docx'}, page_content='# INTRODUCTION TO RAGS ## Analysis vs Analytics Alright! So… Let’s discuss the not-so-obvious differences between the terms analysis and analytics. Due to the similarity of the words, some people believe they share the same meaning, and thus use them interchangeably. Technically, this isn’t correct. There is, in fact, a distinct difference between the two. And the reason for one often being used instead of the other is the lack of a transparent understanding of both')
```

```python
vectorstore_from_dir.add_documents([add_doc])
```
